```python
import pandas as pd
import os
import requests
from bs4 import BeautifulSoup
```

```python
#put in file path that contains list of real estate agents
df = pd.read_csv(r"C:\Users\THW_9\Desktop\deskyop\ResearchAugust\df.csv")
```

```python
# iterate over file and searches on google the agent's name and registration number to get the agent's photo
for i in range(29000,len(df)):
    print(i)
    word = df.loc[i,'registration_no'] + " "+ df.loc[i,'salesperson_name'] + " " + "Real Estate PropertyGuru Agent"
    print(df.loc[i,'salesperson_name'])
    print(word)
    url = 'https://www.google.com/search?q={0}&tbm=isch'.format(word)
    content = requests.get(url).content
    soup = BeautifulSoup(content,'lxml')
    images = soup.findAll('img')
    if len(images)<2:
        print('fail')
    else:
        response=requests.get(images[1].get('src'))
        print(response)
        filepath_1=os.path.join(r"C:\Users\THW_9\Desktop\deskyop\ResearchAugust\Untitled Folder\StoragePic",df.loc[i,'registration_no']+"_"+df.loc[i,'salesperson_name'].replace('/','')+".jpg")
        with open(filepath_1, "wb") as f:
            f.write(response.content)
                            
```
